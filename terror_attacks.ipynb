# Data Analysis using central tendencies of measurement

### ANALYSIS ON TERRORIST ATTACK COLUMN:

---------------

```python
df['Terrorist attacks'].mean()
```

The mean provides a measure of central tendency, indicating the center or average of a distribution. It is commonly used to summarize a dataset with a single representative value. It is calculated by summing up all the
values in a dataset and then dividing the sum by the number
of data points.

The mean of 76.83 in the context of terrorist attacks from 1970 to 2021
suggests the average number of terrorist attacks per year over that time
period.

```python
df['Terrorist attacks'].median()
```

Given that the distribution of terrorist attacks is right-skewed with potential outliers (some years may have significantly more attacks), the median may be a more robust measure.

```python
max=df['Terrorist attacks'].max()
max
```

Hence the maximum number of attacks which took place are 16820

```python
df['Terrorist attacks'].max()
df['Terrorist attacks'].min()
range=df['Terrorist attacks'].max()-df['Terrorist attacks'].min()
range
```

The range is a measure of the spread or dispersion of a set of data. It is
calculated as the difference between the maximum and minimum values
in a dataset.

***CONCLUSION***: A range of 16820 in the context of terrorist attacks refers to the difference
between the highest and lowest numbers of terrorist attacks recorded
over a specified period.

```python
df['Terrorist attacks'].var()
```

In statistics, variance is a measure of the spread or dispersion of a set of
values in a dataset. It quantifies how far each data point in the set is from
the mean (average) and provides an indication of the variability or
"scatter" of the data.

***CONCLUSION:*** A variance of 283,381.69 in the context of terrorist attacks from 1970 to
2021 indicates the degree of dispersion or spread of the number of
terrorist attacks per year over that time period.


```python
df['Terrorist attacks'].std()
```

In statistics, the standard deviation is a measure of the amount of
variation or dispersion in a set of values. It provides a way to quantify how
much individual data points in a dataset differ from the mean (average) of
the dataset.
A low standard deviation indicates that the data points tend
to be close to the mean, while a high standard deviation indicates that the
data points are spread out over a wider range of values.

***CONCLUSION:*** Standard deviation of 532.33 suggests that the number of Terrorist attacks
has fluctuated or varied significantly from year to year over the specified
time period.

## ANALYSIS ON AGE GROUPS

```python
age_columns = ['Death Age 100+', 'Death Age: 51-99', 'Death Age : 21-50', 'Death Age : 11-20', 'Death Age : 6-10', 'Death Age :  1-5']

# Strip whitespaces from column names
df.columns = df.columns.str.strip()

# Check the exact column names in the DataFrame
print(df.columns)

# Create a new column 'Age Group' by combining all age group columns
df['Age Group'] = df[age_columns].idxmax(axis=1)

# Group by 'Age Group' and calculate the mean, median, and range
grouped_age_data = df.groupby('Age Group')['Terrorist attacks'].agg(['mean', 'median', lambda x: x.max() - x.min()]).reset_index()

# Rename the range column for clarity
grouped_age_data.rename(columns={'<lambda_0>': 'Range'}, inplace=True)

# Display the resulting grouped DataFrame
print(grouped_age_data)
```

A new column, 'Age Group,' was created to identify the age group with the highest number of terrorist attacks for each observation.

The dataset is grouped by the 'Age Group' column.
Calculations are performed to determine the mean, median, and range of terrorist attacks within each age group.

## ~Year Vs Deaths(age-wise):A Stacked Bar plot representation

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='Year',y='Terrorism deaths',palette='inferno')
plt.title('Stacked barplot:Year and Deaths(1-100+)')
plt.xlabel('Year')
plt.ylabel('Terrorism Deaths')
plt.xticks(rotation='vertical')
plt.show()
```

## ~Pie Chart: To show the distribution of total deaths in Age Groups in the year 2014

----------

```python
import matplotlib.pyplot as plt
df.columns = df.columns.str.strip()

selected_year = 2014
selected_year_data = df[df['Year'] == selected_year]
total_deaths_by_age = selected_year_data[['Death Age 100+', 'Death Age: 51-99', 'Death Age : 21-50',
                                          'Death Age : 11-20', 'Death Age : 6-10', 'Death Age :  1-5']].sum()

plt.figure(figsize=(20, 20))
plt.pie(total_deaths_by_age, labels=total_deaths_by_age.index, autopct='%1.1f%%',
        startangle=140, colors=plt.cm.Paired.colors)
plt.title(f'Distribution of Total Deaths by Age Groups in {selected_year}')
plt.show()
```

### Conclusion:

#### Pie Chart provides a visual representation of the proportional contributions of the various categories to the whole. We can infer that the number of infant deaths is much higher compared to other age groups and whicb is approximately 8 to 10 times higher.

----------------

## ~Heatmap: Attacks and Deaths Across Age Groups and Years

---------

```python
heatmap_data = df.pivot_table(index='Year', values=['Death Age 100+','Death Age: 51-99','Death Age : 21-50',
       'Death Age : 11-20', 'Death Age : 6-10', 'Death Age :  1-5'], aggfunc='sum')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='GnBu', annot=True, fmt='g', linewidths=.5, vmin=0,vmax=25000)
plt.title('Heatmap: Attacks and Deaths Across Age Groups and Years')
plt.xlabel('Age Groups')
plt.ylabel('Attack Year')
plt.show()
```

### Conclusion:

#### Heatmap visually represents the intensity or magnitude of values with a matrix. Via the use of colorcoding we can see the specific values of deaths between different age groups. As mentioned previously the infant age group has the highest number of deaths which concludes that younger segment of the population is far more vulnerable to terror attacks compared to the rest.

## ~A Boxplot presenting Terrorist Attacks and Terrorism caused Deaths

------

```python
df.columns = df.columns.str.strip()

data_for_boxplot = df[['Terrorist attacks', 'Terrorism deaths']]
sns.set(style="whitegrid")
fig2, ax2 = plt.subplots(figsize=(10, 6))
boxprops = dict(linewidth=2, color='darkblue')
medianprops = dict(linewidth=2, color='orange')
whiskerprops = dict(linewidth=2, linestyle='--', color='gray')
capprops = dict(linewidth=2, linestyle='--', color='gray')
ax2.boxplot(data_for_boxplot, labels=['Terrorist Attacks', 'Terrorism Deaths'], showfliers=False,
           boxprops=boxprops, medianprops=medianprops, whiskerprops=whiskerprops, capprops=capprops)

ax2.set_ylabel('Count')
ax2.set_title('Box Plot of Terrorist Attacks and Terrorism Deaths')

plt.show()
```

### Conclusion:

#### A Boxplot provides a concise summary of the distribution of a total dataset. It offers insight into its central tendency spread and presence of outliers. For the Terrorist Attacks and Terrorist Deaths we can interpret the median, 4th quatrtile and higher extreme value. The range of terrorist attacks is comparatively higher to the number of terrorist deaths.

--------------------

## ~A time series line plot: Terrorist Death types

------------

```python
df.columns = df.columns.str.strip()

years = df['Year']
armed_assault = df['Terrorist Death Type : Suicide']
bombing_or_explosion = df['Terrorist Death Type : Killed']

plt.figure(figsize=(12, 6))
plt.plot(years, armed_assault, label='By Suicide', marker='o', linestyle='-.', color='red', alpha= 1)
plt.plot(years, bombing_or_explosion, label='Killed', marker='d', linestyle=':', color='blue', alpha= 0.4)

plt.xlabel('Year')
plt.ylabel('Terrorist Death')
plt.title('A time series line plot:Terrorist Death Type')

plt.legend()
plt.grid(True)
plt.xticks(range(min(years), max(years)+1, 1), rotation=50, ha='right')
plt.show()
```

### Conclusion:

#### Line plot illustrates a clear pattern in the data this visual representation enchances our understanding of the topic. In our lineplot blue markers represent number of terrorists killed and red marker represemts the terrorists who commitedsuicide.
